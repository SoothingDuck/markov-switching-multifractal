# Abstract

The purpose of this paper is to outline a research agenda that confronts the question: Does bringing more information into the prediction-problem improve prediction?  Over the course of two papers I propose to investigate the impact of more information upon volatility forecasting, with the aim to provide insight.  I shall do this in two ways: firstly, by observing an economy comprising high frequency data; and secondly by using Machine Learning to bring a much broader range of data features and potential predictors to the same economy.  My first paper shall apply the Markov Switching Multifractal of Calvet & Fisher to tick-by-tick FX data comprising in excess of 10 million observations for each of the most liquid currency pairs.  I shall also compare the performance of the MSM to a constant forecast and other standard volatility models, including the GARCH(1,1), MS GARCH, multivariate GARCH, and long-memory stochastic volatility model.  In my second paper I shall use Machine Learning to apply high dimensional non-linear methods to the same volatility forecasting problem and compare Machine Learning’s predictive performance to that of MSM under the conditions established in the first paper.  This research shall extend and complement the literature on multifractal analysis, regime switching, multi-frequency volatility risk, high frequency volatility forecasting, and the nascent literature on the application of Machine Learning methods to problems in finance, and to volatility forecasting in particular.

Finance researchers have shown for many decades that asset returns measured at high frequencies are consistently far from normal, exhibit excess kurtosis, and have a strong tendency to stay either very close to the mean or far from it.  This non-normality diminishes when returns are measured over longer time intervals.  Conditioning information can be used to predict both first and second moments of asset returns.  The predictability in first moments is elusive, as might be expected if indeed markets are efficient.  Conversely however, the predictability in second moments has been shown to be stronger.   Literature has documented that conditional volatility can move suddenly but also varies persistently, so that time-varying volatility is apparent at lower frequencies (for example, over many years) and also at much higher frequencies such as intra-day.  

As a result several classes of models for volatility of returns have been written, accounting for many hundreds of volatility forecasting models .  Each class however exhibits limitations.  In very succinct terms, the (G)ARCH framework fails to offer an integrated explanation of return phenomena at different frequencies.  Stochastic volatility models are subject to limitations arising from jumps.  Discrete-state regime-switching models allow stochastic volatility to jump .  Regime-switching models have proven to be tractable, but the observer of financial markets sees many possible states of volatility.   This has tended to limit the use of regime-switching models because the number of parameters in regime increases with the square of the number of states, so that models become unusable very quickly as new states are added .

### Multifractal Modeling

It is more than half a century since Benoit Mandelbrot discovered that financial returns may strongly depart from Gaussian and advocated the use of self-similar Levy-stable processes for modeling market fluctuations, these two insights in turn marking the introduction of fractal methods in finance.  Fractal tools such as fat tails, fractional integration, and multifractal scaling are now broadly used and often combined with other approaches.  They have also given rise to popular hybrid models such as fractionally integrated GARCH  or long-memory stochastic volatility . 

An outline of the evolution of multifractal modeling in the literature is shown in tabular form blow.  It can be seen that over last twenty years, fractal research in finance has centered on the development of multifractal models of returns, which can jointly capture fat tails, long-memory volatility persistence, multifractal moment scaling, and nonlinear changes in the distribution of returns observed over various horizons.  Multifractal modeling has extended to the commodity markets over the last five years, where large fluctuations in price volatility are particularly well suited to multifractal models such as the MSM ability to capture stylized facts of multiscaling, long memory, and the structural breaks in volatility that are particular features of commodity markets .  Multifractal measures can now be constructed dynamically through time , whilst several classes of multifractal diffusions are now available .  MSM shows that multifractality can be generated by a Markov process with multiple components, each of which has its own degree of persistence.  MSM also permits the application of efficient statistical methods, such as likelihood estimation and Bayesian filtering, to a multifractal process.  Incorporating multifractal risk into a pricing model generates multifractal jump-diffusions .  These innovations provide new intuitions about multifractal behavior in economic phenomena.

Fractal research in finance has centered on the development of multifractal models of returns, which can jointly capture fat tails, long-memory volatility persistence, multifractal moment scaling, and nonlinear changes in the distribution of returns observed over various horizons. Multifractal models capture these empirical regularities with a remarkably small number of parameters.  The MSM treats volatility as the product of a vector of state components that follow independent regime-switching processes.  The components have identical marginal distributions and heterogeneous durations.  Thus, some may switch on average only once every several years or decades, while others can have durations measured in days or less.  Scale invariance properties permit modelling of all price variations using a single data-generating mechanism. As a consequence, MSM is extremely parsimonious. A small number of well-identified parameters, combined with testable assumptions on scale-invariance, specify price dynamics at all timescales.  The Markov-switching approach to multifractal modeling thus provides a parsimonious and rigorous way of modeling a financial time series.  It provides a powerful, yet arguably under-researched, way to connect phenomena that are generally modeled separately.

Multifractal research in finance remains a young and field, and many challenges remain.  The statistical methodology can be improved to incorporate finer features of financial returns . Improvements in statistical inference are possible, for instance by using different distributions for M, by exploring different transition probability specifications, and by modifying the estimation method .  The integration of fractal risk into pricing models, whilst nascent, is illustrated by work on option pricing and the term structure of interest rates .  The applications of multifractal techniques to finance nonetheless remain nascent. 

Of particular relevance, from the perspective of this research, the multifractal behavior of financial returns remains unexplained and invites the exploration of the economic mechanisms producing self-similar regime-switching in financial volatility.  Theory suggests that the predictive performance of the MSM when high frequency data is used should be good because a larger number of volatility components allows each component to do less work in explaining aggregate volatility variations.  This research addresses the question: Does bringing more information into the prediction-problem improve prediction?  It therefore extends the literature on multifractal modeling in two ways: firstly, I shall forecast volatility based upon a much higher number of observations than generally observed in the literature using MSM; and secondly, the use of ML in the second paper to forecast volatility and examine the impacts of high dimensionality and nonlinearities, and subsequent comparison with MSM, may provide additional insights into the multifractal behavior of financial information.

#### MSM

The MSM of Calvet & Fisher (i.e. 2001, 2002a, 2002b, 2004, 2006, 2007, 2008a, 2008b) is regarded as a major advance in the modern literature on financial econometrics (Campbell, 2008). MSM is a stochastic volatility model and assumes that return volatility is a multifractal system.  The assumption of multifractality is consistent with key stylized facts of financial markets returns, including fat tails, scaling, volatility clustering, and long memory.   

Depending upon the specification, MSM may have a finite or infinite number of states.  Maximum likelihood estimation may be used to estimate the parameters of the MSM (i.e. Calvet & Fisher (2001, 2002a, 2004, 2007), Lux(2006, 2008), Lux & Kaizoji (2007), Liu, Matteo & Lux (2007), Segnon, Lux & Gupta (2015)).  However, the maximum likelihood approach is computationally expensive and becomes intractable as the number of states are increased, as in this research to enhance predictive capability.  This links the MSM to the LSTM RNN in several interesting ways, which I shall address subsequently in LSTM-specific sections, in particular in paper 2.

##### MSM model

The MSM model is sometimes denoted $MSM(\bar k)$ where $\bar k$ is a parameter representing the number of multipliers $M_{i,t}\ \ i = 1,...,\bar k$. 

$M_t$ may therefore be defined as a vector of these multipliers at time $t$, treated as a state at $t$, and written:

$$
\begin{align}
   M_t = (M_{1,t},...,M_{\bar k,t})\in \mathbb{R}^{\bar k}_+.
\end{align}
$$

Each component of $M_t$ has the same marginal distribution $M(\theta)$ but differs in its switching frequency, which is given by the exogenous parameter $\gamma \equiv (\gamma_1,\gamma_2,...,\gamma_{\bar k})$. The dynamics of multipliers may therefore be written:

$$
\begin{align}
    M_{k,t} = 
    \begin{cases}
        m \sim M(\theta) & \text{with probability $\gamma_k$}\\
        M_{k,t-1} & \text{with probability $1-\gamma_k$}.
    \end{cases}
\end{align}
$$

Where the exogenous switching probability $\gamma_k$ is written:

$$
\begin{align}
    \gamma_k = 1 - (1-\gamma_1)^{b^{k-1}}.
\end{align}
$$

Due to this construction, multipliers are mutually independent of each other and the stochastic volatility may be considered as a function of the vector $M_t$ with a first-order Markov property.  This may be written:

$$
\begin{align}
    \sigma(M_t) \equiv \bar \sigma * \Big(\prod_{i=1}^{\bar k} M_{k,t}\Big)^{\frac{1}{2}},
\end{align}
$$

Where $\bar \sigma$ is a positive constant to be estimated.

Returns $r_t$ may be written:

$$
\begin{align}
    r_t = \sigma(M_t)\epsilon_t \ \ \ \text{where $\epsilon_t \sim N(0,1)$}.
\end{align}
$$

Which is a realization of the state $\sigma(M_t)$ with Gaussian error.

The full parameter vector may therefore be written $\psi \equiv (b, \theta,\gamma_{1},\bar \sigma)$ and the marginal distribution $M(\theta)$ may be any distribution with positive support providing $\mathbb{E}(m) = 1$.

Using the simplest specification of $M(\theta)$ where $m$ takes just two values with equal probability, the transition matrix of the state $M_t$ is $2^{\bar k}*2^{\bar k}$. 

My research focuses upon the impact of more information on volatility prediction performance in settings of very large amounts of UHF data.  I therefore use this specification and specify $\theta = m_0$ following Calvet and Fisher (2004), where $M(m_0)$ is a binomial distribution which yields either $m_0$ or $m_1 = 1-m_0$ with equal probability.

##### Implementation of MSM in Python

My starting point writing the MSM in Python was the original MatLab code of Calvet and Fisher (2004), available at RunMyCode.  It can be found __[here](http://www.runmycode.org/companion/view/31)__.  The conversion from MatLab to Python presented some challenges but was generally relatively straightforward.  The initial model thus written however ran slowly on a high specification CPU, particularly with higher $\bar k$ settings, using HSBC FX repo data (inside the HSBC firewall) and, subsequently, the benchmark datasets (on a new higher specification machine outside of the firewall). Use of a Graphic Processing Unit (`GPU`) to enhance computing power inside HSBC had limited effect because the associated Compute Unified Device Architecture (`CUDA`) and associated Python libraries such as cuPY were not available within the firewall.  After moving completely outside of the HSBC firewall following rejection of the FX repo data as a data source for this research, ensuing work with the very large HistData.com EURUSD UHF datasets indicated that neither estimation nor n-step volatility forecast-based performance assessment would be possible, especially with the targeted full-scope UHF datasets of 6.2 million observations (19 years 1-minute bar data) and 128 million observations (11 years tick-by-tick data) respectively.  In fact it was not possible to run the Python model with relatively small subsets of the UHF data, for example, 1 year of 1-minute bar and 1 year of tick-by-tick data respectively.

I therefore re-wrote a number of the MSM Python files and introduced several optimizations, in order to improve performance.  I introduced a particle filter for simulated likelihood approximation, state estimation by filtering, and as an additional comparative benchmark for performance assessment via the n-step ahead volatility forecasts.  I further optimized the core MSM and the particle filter using methods of multiprocessing and memoization, for further performance efficiency.  These innovations take place within the context of the wider search for insight regarding the impact of more information on volatility forecasting, as better performing models allow more data to be brought into the forecasting problem setting but also shed light upon the ways in which more data potentially impacts the forecast.

This Empirical Study document is written as a Jupyter Notebook, which allows my code to be shown in conjunction with the model estimation, volatility forecasting performance assessment, statistical analysis, and visualization that it performs. ##### MSM estimation

$MSM(\bar k)$ estimation requires presentation to the model of a time series of returns with mean zero and for the number of multipliers $\bar k$ to be set.  The model then estimates parameters $\psi \equiv (b,m_0,\gamma_{\bar k},\bar \sigma,)$. 

**Starting value search**

My starting_val.py function is a Python translation of the original MatLab code. It finds starting values for parameters $\psi$ to be used in maximum likelihood estimation. Firstly, the parameters except $m_0$ are given several candidate values. Then I run an optimization `scipy.optimize.fminbound()` to maximize likelihood with respect to $m_0$ bounded between 1.2 and 1.8, following the MatLab original, for all combinations of other parameters.  I thus perform a grid search.

__Likelihood functions__

I have implemented the likelihood functions in three ways: starting value search, MLE, and state estimation. For this reason, different set of variables are returned depending upon the input.

When the input for the set of parameters is not a vector (i.e. the starting values for the parameters have not been passed as inputs), the function only returns the likelihood; thus, in this case, the function performs a starting value search. Once determined, starting values are supplied to the function minimizer. Then the likelihood function is invoked again still limiting the number of outputs to likelihood only, to perform a global minimum search with `scipy.optimize.basinhopping()`.

To compute exact likelihood, I first define the conditional probability of $M_t$ being at one of its $2^{\bar k}$ states given the return at time t as: 

$$
\begin{align}
\Pi_t^j = P(M_t = m^j|r_t)
\end{align}
$$

Then I observe the return at time $t+1$ and update the distribution of the above equation (6) probability $\Pi_t = (\Pi^1_t,...,\Pi^{2^{\bar k}}_t)$ using the Bayesian update:

$$
\begin{align}
\Pi_{t+1} \propto f(x_{t+1}|M_{t+1} = m^j)\Pi_tA
\end{align}
$$

where A is the $2^{\bar k}*2^{\bar k}$ transition matrix of states.

The log likelihood function up to time $T$ is then written: 

$$
\begin{align}
\log(L(r_1,...,r_T;\psi) = \sum_{t=1}^T \log(f(x_{t}|M_{t} = m^j)\Pi_{t-1}A)
\end{align}
$$

##### Volatility forecasting using the MSM

When the number of inputs parameter `narg` is set to a value other than 1, the likelihood function returns the data likelihood plus daily likelihood, the posterior of $\Pi_T$, along with the daily filtered state estimate (i.e. $\Pi_{t-1}A$ at time $t$).

Parameters are then estimated by the likelihood function for the first $T - nsteps$ observations, with forecasts for the final $nsteps$.

#### Particle Filter

Confronting my Python implementation of MSM with, firstly, HSBC FX repo trades and market datasets, then subsequently my benchmark datasets and sub-sets of UHF data,  I observe the maximum likelihood approach to be computationally expensive, so that it becomes intractable as $\bar k$ increases and/or the size of the dataset confronted to the model increases. In addition to my code modification and optimization, I therefore also apply a particle filter to approximate the likelihoood and simulate volatility predictions.  Lux (2018) observes that the use of a particle filter to compute a simulated maximum likelihood can be computationally more efficient.  My implementation in Python of the particle filter follows Liu & West (2001) and Greve, et al (2018).

As $\bar k$ increases, obtaining the transition matrix in particular becomes prohibitively expensive in terms of computation.  Optimization of the MSM Python code significantly improves model performance, but the introduction of UHF data at $\bar k$ settings of 10 (a setting used for much of my empirical work), led again to a computational impasse.  For example, an initial estimation of MSM with likelihood, setting $\bar k = 10$ and using 1 year (2018) tick-by-tick data (18 million observations), failed even after running for over 100 hours. The particle filter, or sequencial monte carlo, seeks to address this problem.

The particle filtering approximates $\Pi_{t+1}$ with two stage $M_{T+1}$ sampling. Let F be a function that maps $\mathbb{R}$ to $\mathbb{R}_{+}^{\bar{k}}$ and let d be the number of possible values that $m^{j}$ can take and $R_{t+1} \equiv \{r_s\}_{s=1}^{t+1}$. 

Let $h(m) \equiv P(M_{t+1}|R_t)$, so that:

$$
\begin{align}
E[F(M_{t+1})|R_{t+1}] &= \sum_{j=1}^d P(M_{t+1}=m^j|R_{t+1})F(m^j) \\
&= \sum_{i=1}^d h(m^j)\frac{P(M_{t+1}=m^j|R_{t+1})}{h(m^j)}F(m^j)
\end{align}
$$

Because,

$$
\begin{align}
h(m^j) = P(M_{t+1} = M^j|R_t)
\end{align}
$$

By the convergence of the Monte Carlo Integration:

$$
\begin{align}
E[F(M_{t+1})|R_{t+1}] &\approx \frac{1}{B}\sum_{b=1}^B \frac{P(M_{t+1}=\hat{M}_{t+1}^{(b)}|R_{t+1})}{h(\hat{M}_{t+1}^{(b)})} F(\hat{M}_{t+1}^{(b)})
\end{align}
$$

Which, effectively, is Bayesian importance sampling, because:

$$
\begin{align}
\frac{P(M_{t+1}=\hat{M}_{t+1}^{(b)}|R_{t+1})}{Bh(\hat{M}_{t+1}^{(b)})} = \frac{f_{r_{t+1}}(r_{t+1}|M_{t+1}=\hat{M}_{t+1}^{(b)})}{Bf_{r_{t+1}}(r_{t+1}|R_t)}
\end{align}
$$

Moreover, the denominator can be approximated as follows

$$
\begin{align}
f_{r_{t+1}}(r_{t+1}|R_t) &\approx \frac{1}{B}\sum_{b=1}^B f_{r_{t+1}}(r_{t+1}|\hat{M}_{t+1}^{(b)})
\end{align}
$$

Therefore:

$$
\begin{align}
E[F(M_{t+1})|R_{t+1}] \approx \sum_{b=1}^B \frac{f_{r_{t+1}}(r_{t+1}|M_{t+1}=\hat{M}_{t+1}^{(b)})}
{\sum_{b=1}^B f_{r_{t+1}}(r_{t+1}|\hat{M}_{t+1}^{(b)})} F(\hat{M}_{t+1}^{(b)}) \\
\end{align}
$$

$$
\begin{align}
= \sum_{b=1}^B W_b F(\hat{M}_{t+1}^{(b)})
\end{align}
$$

Thus the expectation $E[F(M_{t+1})|R_{t+1}]$ can be approximated with the weighted average of the statistics of drawn samples, weighted by the likelihoods of the returns. This resembles the form of approximation using Bayesian importance sampling. Since the function $F$ is defined as a mapping from the real number space to the positive real value space, $F$ may be replaced with the state conditional probability $\Pi_{t+1}$.

It follows therefore that the particle filter approximates $\Pi_{t+1}$ with re-sampled $\{\hat{M}_{t+1}^{(i)}  \}_{i=1}^B$.

##### Implementation of the particle filter in Python

At fixed time $t$ I draw: $\{M_{t+1}^{(i)}\}_{i=1}^B$ using $P(M_{t+1}^{i}|M_{t}^{i})$, being one row of a transition Matrix A.  Then I re-weight the probability of draws using importance sampling and re-draw $\{\hat{M}_{t+1}^{(i)}\}_{i=1}^B$ using the re-weighted probabilities.

I assume that $\{M_{t}^{(i)}\}_{i=1}^B$ has been independently drawn from $\Pi_t$, so that $\Pi_{t+1}$ can be updated given a new return $r_{t+1}$ thus: 

$$
\begin{align}
$\Pi_{t+1}^j \propto f_{r_{t+1}}(r_{t+1}|M_{t+1}=m^j)\sum_{i=1}^d P(M_{t+1}=m^j|M_t=m^i)\Pi_t^i$  
\end{align}
$$

As a first step therefore I draw $M_{t+1}^{(i)}$ given $M_{t}^{(i)}$ using the one-step ahead Markov chain property of $M_t$.  I repeat this B times in order to obtain the first stage of $\{M_{t+1}^{(i)}\}_{i=1}^B$. I then take the weighted average of states to obtain a filtered estimate of $M_{t+1}$ using the information up to time $t$. 

With the observation at time $t+1$, I may update $M_{t+1}$ incorporating the information of the return at $t+1$, as follows:

For a fixed sample $b$, compute

$$
\begin{align}
$P(q=b) = \frac{f_{r_{t+1}}(r_{t+1}|M_{t+1}=M_{t+1}^{(b)})}{\sum_{a=1}^B f_{r_{t+1}}(r_{t+1}|M_{t+1} = M_{t+1}^{(a)})}$.
\end{align}
$$

Then draw B samples of $M_{t+1}$ from $\{M_{t}^{(i)}\}_{i=1}^B$ with corresponding weighted probabilities.  Define them as $\{\hat{M}_{t+1}^{(i)}\}_{i=1}^{B}$, following the conditional distribution $\Pi_{t+1}$.

##### Forecasting using the particle filter

At fixed time $T$, I draw $\{M_{T+1}^{(b)}\}_{b=1}^B$ using $P(M_{T+1}^{(b)}|M_{T}= M_{T}^{(b)})$. I then calculate the probability of draws using importance sampling and obtain from the probabilities $\tilde{M}_{T+1}$.  

As a first step I therefore draw $M_{T+1}^{(b)}$, again using the one-step forward Markov chain property of $M_t$. After repeating this sampling for $\forall b$, I am able to obtain temporary samples of $\{\hat{M}_{T+1}^{(i)}\}_{i=1}^B$. 

In the second stage of sampling, I calculate the weight $P(q=b) = \frac{f_{r_{T}}(r_{T}|M_{t+1}=M_{T+1}^{(b)})}{\sum_{a=1}^B f_{r_{T}}(r_{T}|M_{T+1} = M_{T+1}^{(a)})}$ for $\forall b$. 

At this point I deviate from the particle filter sampling $M_t$ up to $t=T$, where the likelihoods of one period forward returns are used to calculate the weights of each sample; instead I use the likelihood of the return at current time $t=T$ because I do not have $r_{T+1}$ at $t=T$.

I conclude by obtaining the weighted average $\tilde{M}_{T+1} = \sum_{b=1}^B p(q=b)\hat{M}_{T+1}^{(b)}$ with samples' corresponding probabilities.

Since the likelihood of the return at $t=T+1$ is determined by $M_{T+1}$, predicting the stochastic volatility index $M_{T+1}$ is equivalent to implementing a prediction on the return $r_{T+1}$

#### GARCH (1,1)

For benchmarking, I use the GARCH(p,q) model `GARCH(1,1)` of Bollerslev (1986). I use a standard Python implementation of GARCH(1,1) with a constant mean, following __[Sheppard (2018)](https://github.com/bashtage/arch)__.

I describe the GARCH(1,1) model as follows:

$$
\begin{align}
r_t        & =    & \mu + \epsilon_t \\
\epsilon_t & =    & \sigma_t e_t \\
\sigma^2_t & =    & \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma^2_{t-1} \\
e_t        & \sim & N(0,1)
\end{align}
$$

##### Implementation of GARCH (1,1) in Python

My core GARCH model contruction in Python following Sheppard (2018) accepts the following parameters:

- y == The dependent variable.
- mean == Name of the mean model, being one of ‘Constant’, ‘Zero’, ‘ARX’ and ‘HARX’.
- lags == A scalar indicating lag length.
- vol == Name of the volatility model, being one of ‘GARCH’ (default), ‘ARCH’, ‘EGARCH’, ‘FIARCH’ and ‘HARCH’.
- p == Lag order of the symmetric innovation.
- o == Lag order of the asymmetric innovation.
- q == Lag order of lagged volatility or equivalent.

##### Forecasting using the GARCH(1,1)

For n-step volatility forecasts using the GARCH(1,1) model I estimate the model using the dataset minus n-steps, which I denote `train`, and then produce forecasts for the final n-steps, which I denote `test`.

My forecast then evolves in terms of the squares of the model residuals, exploiting the relationship $E_t[\epsilon_{t+1}^2] = \sigma_{t+1}^2$ to recursively compute the forecast.

The variance forecasts are constructed for the conditional variances as

$$
\begin{align}
\sigma^2_{t+1} & = & \omega + \alpha \epsilon_t^2 + \beta \sigma^2_{t-1} \\
\sigma^2_{t+h} & = & \omega + \alpha  E_{t}[\epsilon_{t+h-1}^2] + \beta E_{t}[\sigma^2_{t+h-1}] \, h \geq 2 \\
                  & = & \omega + \left(\alpha  + \beta\right) E_{t}[\sigma^2_{t+h-1}] \, h \geq 2
\end{align}
$$

#### LSTM RNN

In my second paper I introduce a machine learning model, the Long Short Term Memory Recurrent Neural Network (`LSTM RNN` or `LSTM`), originally reported by Hochreiter & Schmidhuber (1997), but in my second paper following the architecture of Xiong, Nichols, & Shen (2016).

Long Short-Term Memory (LSTM) networks are recurrent neural networks equipped with a special gating mechanism that controls access to memory cells (Hochreiter & Schmidhuber, 1997). Since the gates can prevent the rest of the network from modifying the contents of the memory cells for multiple time steps, LSTM networks preserve signals and propagate errors for much longer than ordinary recurrent neural networks. By independently reading, writing and erasing content from the memory cells, the gates can also learn to attend to specific parts of the input signals and ignore other parts. These properties allow LSTM networks to process data with complex and separated interdependencies.

The application of neural networks in canonical finance settings is a rapidly advancing area of knowledge (Kelly, 2019). The seminal LSTM paper of Hochreiter & Schmidhuber (1997) is considered a milestone in the timeline of machine learning.  More recently, in the current decade in particular, recurrent neural networks and LSTMs specifically have seen broadening interest in finance settings - i.e Heaton et al., 2016a; Sirignano, 2016; Xiong, Nichols, & Shen (2016), He (2016), Feng et al., 2018; Gu, Kelly & Xiu, 2018; Verstyuk, 2019).  I have selected the LSTM RNN to complete my model line up because it has been shown to be well-suited to prediction based on time series data, both within and outside of finance settings.

An LSTM module (or cell) has 5 essential components which allows it to model both long-term and short-term data.  They are:

**`Cell state` ($c_t$)**

The cell state represents the internal memory of the LSTM cell and stores both short term memory and long-term memories.

**`Hidden state` ($h_t$)**

The hidden represents output state information that is calculated with regard to current input, previous hidden state, and current cell input. Additionally, the hidden state is enabled to decide to only retrieve the short or long-term or both types of memory stored in the cell state.

**`Input gate` ($i_t$)**

The input gate computes how much information from current input flows to the cell state.

**`Forget gate` ($f_t$)**

The forget gate computes how much information from the current input and the previous cell state flows into the current cell state.

**`Output gate` ($o_t$)**

The output gate computes how much information from the current cell state flows into the hidden state, so that if needed, the LSTM cell can select either long-term memories, or both short-term memories and long-term memories.

### Optimization

#### Short summary of steps taken to optimize the set up

To strengthen performance of the MSM in the presence of very large UHF datasets I firstly implemented a more efficient computation of the transition matrix A. Since the dimensions of the matrix are $2^{\bar k}*2^{\bar k}$, the computational load of this operation increases exponentially. This of course is the curse of dimensionality cited by Calvet & Fisher (2008a). In order to make the construction of the transition matrix more efficient, compared to the original MatLab implementation for example, I simplified the use of two functions to do this (T_mat_template() and transition_mat()) by using the Python np.fromfunc(). This resulted in a significant improvement in computation speed, particularly with higher $\bar k$ settings. The introduction of  basin-hopping, a stochastic algorithm that finds the global minimum  more efficiently, also significantly improved compute times for the likelihood function.

For insight as well as in search of improved performance, I introduced the particle filter of Liu & West (2001). In my implementation, which follows the architecture of Greve, et al (2018), I need only to have a single row of the transition matrix which corresponds to the probability of moving to $2^{\bar k}$ states given current the state of the sample. In this way, the computational complexity of the procedure becomes linear to $2^{\bar k}$, in comparison to the exact computation by the MSM, which requires the full transition matrix and relative computational complexity of the square of $2^{\bar k}$.

For performance assessment using n-step ahead volatility forecasting, the particle filter reuses the samples from the previous step and assigns a new state depending upon the transition probability.  This is continued until the n-th step is reached. The MSM by contrast computes 1st, 2nd, ...nth power of the transition matrix, which is then applied to the posterior distribution $\Pi_t$.  This is observed in the empirical study to be a much more computationally expensive operation.

My research requires the confrontation of very large quantities of UHF data to the models.  I have yet to deal with the challenges that the purely data centric approach of the LSTM RNN introduces and shall cover this in my next phase of work. Optimization thus far has focused on methodological (i.e. simulation using a particle filter) and code-based optimizations. ##### Performance assessment (n-step volatility forecasts)

I address the question 'does bringing more information into the prediction-problem improve prediction' by comparing relative volatility prediction performance based upon an n-step forward prediction.  I do this firstly with simulated data, and then with data from Calvet & Fisher (2004) in order to replicate their results for the GARCH(1,1) and MSM.

I compare the performance of the GARCH(1,1) of Bollerslev (1986), the univariate binomial MSM of Calvet & Fisher (2004), a simulated MSM forecast using a Liu & West (2001) particle filter, and the LSTM RNN of Verstyuk (2019). For all MSM based methods I take $\bar k = 10$ which is the setting used by Calvet & Fisher in their 2004 paper.  I maintain the $T$ = 7087 setting, for consistency.

For the comparison, I follow Calvet & Fisher (2004) and use Mincer-Zarnowitz OLS (MZ-OLS). 

MZ-OLS performs a simple linear regression of realized volatility (square of returns) to forecast volatility (i.e. the prediction of my models) and conducts diagnostics to provide coefficients for the input and the constant. An ideal unbiased forecast will have a coefficient for the input close  to 1 and that for the constant close to 0. Thus I can compare relative forecasting performance by the respective proximities to these ideal values.


__Re-simulate data using $\bar k = 10$ following Calvet & Fisher (2004)__

```python
kbar = 10
T = 7087        
startingvals = None 
niter_basin = 1         
```

```python
m0 = m0
m1 = 2-m0
g_s = np.zeros(kbar)
M_s = np.zeros((kbar,T))
g_s[0] = 1-(1-g_kbar)**(1/(b**(kbar-1)))
for i in range(1,kbar):
    g_s[i] = 1-(1-g_s[0])**(b**(i))

for j in range(kbar):
    M_s[j,:] = np.random.binomial(1,g_s[j],T)
dat = np.zeros(T)
tmp = (M_s[:,0]==1)*m1+(M_s[:,0]==0)*m0
dat[0] = np.prod(tmp)
for k in range(1,T):
    for j in range(kbar):
        if M_s[j,k]==1:
            tmp[j] = np.random.choice([m0,m1],1,p = [0.5,0.5])

    dat[k] = np.prod(tmp)
dat = np.sqrt(dat)*sig* np.random.normal(size = T)
dat = dat.reshape(-1,1)
dat.shape
```

```python
b = 6
m0 = 1.6
g_kbar = 0.8
sig = 2/np.sqrt(252)    # I divide sigma by sqrt of the number of trading days
kbar = 10
T = 7087        
startingvals = None 
niter_basin = 1      
n_steps = 40
```

Simulate the data:

```python
m0 = m0
m1 = 2-m0
g_s = np.zeros(kbar)
M_s = np.zeros((kbar,T))
g_s[0] = 1-(1-g_kbar)**(1/(b**(kbar-1)))
for i in range(1,kbar):
    g_s[i] = 1-(1-g_s[0])**(b**(i))

for j in range(kbar):
    M_s[j,:] = np.random.binomial(1,g_s[j],T)
dat = np.zeros(T)
tmp = (M_s[:,0]==1)*m1+(M_s[:,0]==0)*m0
dat[0] = np.prod(tmp)
for k in range(1,T):
    for j in range(kbar):
        if M_s[j,k]==1:
            tmp[j] = np.random.choice([m0,m1],1,p = [0.5,0.5])

    dat[k] = np.prod(tmp)
dat = np.sqrt(dat)*sig* np.random.normal(size = T)
dat = dat.reshape(-1,1)
dat.shape
```

I propose to use two set ups: firstly, a single predictor, `simulated volatility`; and secondly, two predictors, being `simulated returns` and `simulated volatility`.  I therefore need to compute `simulated returns` and `simulated volatility` using the simulated dataset `dat`.

Compute $r_t$ and $r_t^2$:

See section 1.1 of Andersen, et al. (2006).  Following equations 1.1 thur 1.11, I compute returns thus

$$
\begin{align}
    r_t = p_t - p_{t-1}
\end{align} 
$$

Where $p_t$ denotes the logarithmic price of an asset, which in this case is the simulated data point in dataset `dat`.

```python
dat = dat[1:]-dat[0:-1]   # Note that I do not take logs here
dat = dat[dat != 0]
dat = dat[:,np.newaxis]
dfDat = pd.DataFrame(dat)
dfDat.columns = ['returns']
```

I compute realized variance thus

$$
\begin{align}
    \sigma_t^2 = \sum_{i=1}^{N} r_t^2
\end{align}
$$

And I note that realized volatility is simply the square root of the realized variance.

```python
dfDat['sigma']=dfDat['returns']**2   # compute sigma
cols = dfDat.columns.tolist()        # switch order of columns to fit x, y for LSTM
cols = cols[-1:] + cols[:-1]
dfDat = dfDat[cols]
dfDat.head()
```

```python
def z_score(x, mean, std):
    return (x - mean.values) / std.values  # testing set.
```

```python
def z_score_inv(x, mean, std):
    return x * std + mean
```

```python
def apply_z_score_to_data_frame(df, mean, std):
    df = df.apply(lambda x: z_score(x, mean, std), axis=1)

    # Tips: Use this to debug.
    # print(df.head())
    # print(df.apply(lambda x: z_score(x, mean, std), axis=1).head())
    # print(mean)
    # print(std)

    return df
```

```python
def apply_delta_t_to_data_frame(df):
    # we only support those values. They are given by the paper.
    delta_t = 3

    # for trends (MEAN)
    out = df.rolling(window=delta_t, center=False).mean()  # correct.

    # for returns (SUM)
    out['returns'] = df['returns'].rolling(delta_t, center=False).apply(np.sum)  # correct

    # for volatility (SQUARE - SUM - SQRT)
    out['sigma'] = df['sigma'].apply(np.square).rolling(delta_t, center=False).apply(np.sum).apply(np.sqrt)  # correct

    out.dropna(inplace=True)
    return out
```

```python
import os

import numpy as np
from tqdm import tqdm

from data_reader import get_data

# PAPER: Specifically, we set up the maximum lag of the LSTM to include 10 successive observations
LSTM_WINDOW_SIZE = 10

# sigma, returns, comput, crcard, invest, bnkrpt are the most useful predictors.
# because we use mask, we start by sigma first then we add returns, then comput and so forth.

PREDICTORS = ['sigma',
              'returns',
              'Trend COMPUT',
              'Trend CRCARD',
              'Trend INVEST',
              'Trend BNKRPT',
              'Trend AIRTVL',
              'Trend ADVERT',
              'Trend AUTO',
              'Trend AUTOBY',
              'Trend AUTOFI',
              'Trend BIZIND',
              'Trend COMLND',
              'Trend CONSTR',
              'Trend DURBLE',
              'Trend EDUCAT',
              'Trend FINPLN',
              'Trend FURNTR',
              'Trend INSUR',
              'Trend JOBS',
              'Trend LUXURY',
              'Trend MOBILE',
              'Trend MTGE',
              'Trend RENTAL',
              'Trend RLEST',
              'Trend SHOP',
              'Trend SMALLBIZ',
              'Trend TRAVEL',
              'Trend UNEMPL']

if 'DEBUG' in os.environ:
    PREDICTORS = PREDICTORS[0:5]
    print('DEBUG we truncate the predictors because we dont consider all the trends.')
    print(PREDICTORS)

INPUT_SIZE = len(PREDICTORS)


def chunker(seq, size):
    return [(seq[pos:pos + size], seq[pos + size:pos + size + 1]) for pos in range(0, len(seq), 1)]


def df_to_keras_format(df):
    keras_x = []
    keras_y = []

    for x, y in tqdm(chunker(df, LSTM_WINDOW_SIZE), 'df_to_keras_format'):
        # print('*' * 80)

        # filter on predictors
        x = x[PREDICTORS]
        y = y[PREDICTORS]

        # print(x)
        # print(y)

        x_new = x.values
        y_new = y['sigma'].values

        if len(x_new) == LSTM_WINDOW_SIZE and len(y_new) == 1:
            keras_x.append(x_new)
            keras_y.append(y_new)

    keras_x = np.array(keras_x)
    keras_y = np.array(keras_y)
    return keras_x, keras_y


def get_trainable_data():
    tr, te, mean, std = get_data()
    print(tr.head())
    print(te.head())

    x_train, y_train = df_to_keras_format(tr)
    x_test, y_test = df_to_keras_format(te)

    return (x_train, y_train), (x_test, y_test), mean, std


if __name__ == '__main__':
    get_trainable_data()
```

```python
import os
from glob import glob

import numpy as np
import pandas as pd

np.set_printoptions(threshold=np.nan)
pd.set_option('display.height', 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


def read_sp_500():
    spy = pd.read_csv('trends/SP500.csv', parse_dates=True, index_col=0)
    spy['Adj Close Log'] = spy['Adj Close'].apply(np.log)
    spy['Adj Close Log Shift 1'] = spy['Adj Close Log'].shift(1)
    spy['returns'] = spy['Adj Close Log'] - spy['Adj Close Log Shift 1']

    print(spy.head())

    spy.dropna(inplace=True)

    spy.drop('Adj Close Log', 1, inplace=True)
    spy.drop('Adj Close Log Shift 1', 1, inplace=True)

    print('')
    print(spy.head())

    spy['u'] = (spy['High'] / spy['Open']).apply(np.log)
    spy['d'] = (spy['Low'] / spy['Open']).apply(np.log)
    spy['c'] = (spy['Close'] / spy['Open']).apply(np.log)

    print('')
    print(spy.head())
    # correct.
    # 0.511*(0.006651--0.001358)^2-0.019*(0.006651*(0.006651+-0.001358)-2*0.006651*(-0.001358))-0.383*0.006651^2
    # ~ 0.00001482322429 [first value]
    spy['sigma'] = 0.511 * (spy['u'] - spy['d']) ** 2 - 0.019 * (
            spy['c'] * (spy['u'] + spy['d']) - 2 * spy['u'] * spy['d']) - 0.383 * spy['c'] ** 2

    print('')
    print(spy.head())

    spy.drop('u', 1, inplace=True)
    spy.drop('d', 1, inplace=True)
    spy.drop('c', 1, inplace=True)

    spy = spy[['returns', 'sigma']]  # only keep returns and volatility

    print('')
    print(spy.head())

    # import matplotlib.pyplot as plt
    # spy['sigma'].plot()
    # plt.show()
    return spy


def process_trend(trend):
    trend_name = trend.split('/')[1].split('.')[0]
    t = pd.read_csv(trend, parse_dates=True, index_col=0)
    t = t[['Close']]
    t.columns = ['Trend {}'.format(trend_name.upper())]
    print('Trend [{0}] processed.'.format(trend_name))
    # print(t.head())
    return t


def read_trends():
    trends = glob('trends/*.csv')
    trends = filter(lambda x: 'SP500' not in x and 'spy' not in x, trends)
    trends = sorted(trends)  # reproducibility.
    assert len(trends) == 27, 'You should have 27 trends. Check here https://finance.google.com/finance/domestic_trends'
    trends_df_list = []
    for i, trend in enumerate(trends):
        if 'DEBUG' in os.environ and i > 2:
            print('DEBUG! WE TRUNCATE TO ONLY THREE TRENDS FOR SPEED')
            break
        trends_df_list.append(process_trend(trend))
    full_trend_df = pd.DataFrame(pd.concat(trends_df_list, axis=1))
    if full_trend_df.isnull().values.any():
        print(full_trend_df[full_trend_df.isnull().any(axis=1)])
        full_trend_df.fillna(method='ffill', inplace=True)

        if full_trend_df.isnull().values.any():
            full_trend_df.fillna(method='bfill', inplace=True)  # we cheat a very bit at the beginning.
            # we fill gap values by the last values. If we remove lines, we will have gaps in our data
            # and it's not going to be cool.

    assert not full_trend_df.isnull().values.any()
    return full_trend_df


def read_all():
    trends = read_trends()
    sp500 = read_sp_500()
    full_data = sp500.join(trends, how='outer')  # correct

    print('-' * 80)
    print(full_data.tail())
    print('-' * 80)

    full_data.dropna(inplace=True)  # correct

    print('-' * 80)
    print(full_data.tail())
    print('-' * 80)
    # print(read_sp_500().join(process_trend(trend), how='outer'))
    return full_data


def split_training_test(df):
    # 19-Oct-2004 to 9-Apr-2012 while the test set ranges from 12-Apr-2012 to 24-Jul-2015
    training_df = df[:'9-Apr-2015']
    testing_df = df['12-Apr-2015':]
    return training_df, testing_df


def z_score(x, mean, std):
    return (x - mean.values) / std.values  # testing set.


def z_score_inv(x, mean, std):
    return x * std + mean


def apply_z_score_to_data_frame(df, mean, std):
    df = df.apply(lambda x: z_score(x, mean, std), axis=1)

    # Tips: Use this to debug.
    # print(df.head())
    # print(df.apply(lambda x: z_score(x, mean, std), axis=1).head())
    # print(mean)
    # print(std)

    return df


def apply_delta_t_to_data_frame(df):
    # we only support those values. They are given by the paper.
    delta_t = 3

    # for trends (MEAN)
    out = df.rolling(window=delta_t, center=False).mean()  # correct.

    # for returns (SUM)
    out['returns'] = df['returns'].rolling(delta_t, center=False).apply(np.sum)  # correct

    # for volatility (SQUARE - SUM - SQRT)
    out['sigma'] = df['sigma'].apply(np.square).rolling(delta_t, center=False).apply(np.sum).apply(np.sqrt)  # correct

    out.dropna(inplace=True)
    return out


def get_data():
    df = read_all()
    df = apply_delta_t_to_data_frame(df)  # try to apply z-score before and after.
    mean = np.mean(df)
    std = np.std(df)
    df = apply_z_score_to_data_frame(df, mean, std)
    tr, te = split_training_test(df)  # we cheat a bit but very little, no problem.
    return tr, te, mean['sigma'], std['sigma']


if __name__ == '__main__':
    get_data()
```

##### LTSM set up

I propose to use two set ups: firstly, a single predictor, `simulated volatility`; and secondly, two predictors, being `simulated returns` and `simulated volatility`.  I therefore need to compute `simulated returns` and `simulated volatility` using the simulated dataset `dat`.

Compute $r_t$ and $r_t^2$:

See section 1.1 of Andersen, et al. (2006).  Following equations 1.1 thur 1.11, I compute returns thus

$$
\begin{align}
    r_t = p_t - p_{t-1}
\end{align} 
$$

Where $p_t$ denotes the logarithmic price of an asset, which in this case is the simulated data point in dataset `dat`.

```python
dat = dat[1:]-dat[0:-1]   # Note that I do not take logs here
dat = dat[dat != 0]
dat = dat[:,np.newaxis]
dfDat = pd.DataFrame(dat)
dfDat.columns = ['returns']
```

I compute realized variance thus

$$
\begin{align}
    \sigma_t^2 = \sum_{i=1}^{N} r_t^2
\end{align}
$$

And I note that realized volatility is simply the square root of the realized variance.

```python
dfDat['sigma']=dfDat['returns']**2   # compute sigma
cols = dfDat.columns.tolist()        # switch order of columns to fit x, y for LSTM
cols = cols[-1:] + cols[:-1]
dfDat = dfDat[cols]
dfDat.head()
```

The above $r$ and $r_t^2$, i.e. columns `returns` and `sigma` are therefore my spot volatility and returns predictors for the LSTM.

Visualize:

```python
n_steps = len(dfDat)
t = np.arange(0, n_steps)
data1 = dfDat['returns']
data2 = dfDat['sigma']

fig, ax1 = plt.subplots()
color = 'tab:green'
ax1.set_xlabel('n-steps')
ax1.set_ylabel('returns', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('sigma', color=color)  
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title("simulated returns and volatility")
plt.show()
```

Define a function to convert a vector of time series into a 2D matrix:

```python
def convertSeriesToMatrix(vectorSeries, sequence_length):
    matrix=[]
    for i in range(len(vectorSeries)-sequence_length+1):
        matrix.append(vectorSeries[i:i+sequence_length])
    return matrix
```

```python
n_steps = 40 
np.random.seed(1234)
sequence_length = 20
vectorSeries = dfDat['sigma']
matrix_vix = convertSeriesToMatrix(vectorSeries, sequence_length)
matrix_vix = np.array(matrix_vix)
print("Data shape: ", matrix_vix.shape)
```

Transform LSTM input because LSTM layer requires shape of (number of samples, the dimension of each element):

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
print(X_train.shape, X_test.shape)
```

##### MZ regression tests

For the comparison, I follow Calvet & Fisher (2004) and use Mincer-Zarnowitz OLS (MZ-OLS). MZ-OLS performs a simple linear regression of realized volatility (square of returns) to forecast volatility (i.e. the prediction of my models) and conducts diagnostics to provide coefficients for the input and the constant. An ideal unbiased forecast will have a coefficient for the input close  to 1 and that for the constant close to 0. Thus I can compare relative forecasting performance by the respective proximities to these ideal values.

#### Log returns

I test for extreme values in the dataset using the same approach and observe none:

```python
preF = dfTick1y.shape
dfTick1y = dfTick1y.drop(dfTick1y[dfTick1y.bid < 0.5*dfTick1y.bid.mean()].index)
dfTick1y = dfTick1y.drop(dfTick1y[dfTick1y.bid > 1.5*dfTick1y.bid.mean()].index)
dfTick1y.dropna(inplace=True)
postF = dfTick1y.shape
print("Rows removed by filtering extreme values: ","\n"
      "Rows, cols pre-filter = ",preF,"\n"
      "Rows, cols post-filter = ",postF)
```

__Compute log returns for the bid, ask, and mid:__

```python
# compute the mid-price
dfTick1y['mid'] = (dfTick1y['bid'] + dfTick1y['ask'])/2.0
```

```python
logA = np.log(dfTick1y['ask'])
logB = np.log(dfTick1y['bid'])
logM = np.log(dfTick1y['mid'])
```

```python
loga = logA-logA.shift(1)
logb = logB-logB.shift(1)
logm = logM-logM.shift(1)
```

```python
dfTick1y['lrAsk'] = loga
dfTick1y['lrBid'] = logb
dfTick1y['lrMid'] = logm
dfTick1y.head()
```
